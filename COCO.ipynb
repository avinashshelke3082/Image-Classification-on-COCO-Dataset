```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import RMSprop
```

```python
img=image.load_img('/content/drive/MyDrive/COCO/classes_split/train/couch/000000000241.jpg')
plt.imshow(img)
```

```python
cv2.imread('/content/drive/MyDrive/COCO/classes_split/train/couch/000000000241.jpg').shape
```

```python
def rgb_to_gray(image):
  # image = tf.image.rgb_to_grayscale(image)
  # image = tf.image.grayscale_to_rgb(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image=np.resize(image,(image.shape[0],image.shape[1],1))
#     plt.imshow(image)
    return image
```

```python
train=ImageDataGenerator(rescale=1/255,
                         preprocessing_function=rgb_to_gray)
validation=ImageDataGenerator(rescale=(1/255),
                              preprocessing_function=rgb_to_gray)
```

```python
train_dataset=train.flow_from_directory('/content/drive/MyDrive/COCO/classes_split/train',
                                        target_size=(256,256),
                                        batch_size=32,
                                        class_mode='categorical')

validation_dataset=validation.flow_from_directory('/content/drive/MyDrive/COCO/classes_split/val',
                                        target_size=(256,256),
                                        batch_size=32,
                                        class_mode='categorical')
```

```python
# print(train_dataset[0][0][1].shape)
plt.imshow(train_dataset[0][0][3])
```

```python
model=tf.keras.models.Sequential([tf.keras.layers.Conv2D(512,(3,3),activation='relu',input_shape=(256,256,3)),
                                tf.keras.layers.MaxPool2D(2,2),
                                tf.keras.layers.Conv2D(256,(3,3),activation='relu',),
                                tf.keras.layers.MaxPool2D(2,2),
                                tf.keras.layers.Conv2D(128,(3,3),activation='relu',),
                                tf.keras.layers.MaxPool2D(2,2),
                                tf.keras.layers.Conv2D(128,(3,3),activation='relu',),
                                tf.keras.layers.MaxPool2D(2,2),
                                
                                tf.keras.layers.Flatten(),
                                tf.keras.layers.Dense(512,activation='relu'),
                                tf.keras.layers.Dense(256,activation='relu'),
                                tf.keras.layers.Dense(3,activation='softmax')])
```

```python
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer= RMSprop(),
              metrics=['accuracy'])
```

```python
plt.imshow(train_dataset[0][0][0])
```

```python
model_fit=model.fit(train_dataset,
                    steps_per_epoch=32,
                    epochs=20,
                    validation_data=validation_dataset
        )
```

```python
train_dataset.class_indices
```

```python
plt.plot(model_fit.history['accuracy'],label='train')
plt.plot(model_fit.history['val_accuracy'],label='test')
```

```python
plt.plot(model_fit.history['loss'],label='train')
plt.plot(model_fit.history['val_loss'],label='test')
```

```python
results=model.evaluate(validation_dataset[0][0],validation_dataset[0][1],batch_size=32)
```

```python
dir_path='/content/drive/MyDrive/COCO/classes_split/test/motorcycle'
cnt=20
for i in os.listdir(dir_path):
  
   img=image.load_img(dir_path+'//'+i,target_size=(256,256))
   plt.imshow(img)
   plt.show()

   x=image.img_to_array(img)
   x= np.expand_dims(x,axis=0)
   images=np.vstack([x])
   val=model.predict(images)
   val=np.argmax(val)
   if val==0:
     print('couch')
   elif val==1:
     print('horse')
   elif val==2:
     print('motorcycle')
   if(cnt==0):
      break
   cnt=cnt-1
```
